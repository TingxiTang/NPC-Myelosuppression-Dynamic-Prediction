```python
import pandas as pd
```

```python
AllDatasets = pd.read_parquet("./AllDatasets.parquet")
X_pros = pd.read_parquet("./X_pros.parquet")
y_pros = pd.read_parquet("./y_pros.parquet")
X_test = pd.read_parquet("./X_test.parquet")
y_test = pd.read_parquet("./y_test.parquet")
X_train = pd.read_parquet("./X_train.parquet")
y_train = pd.read_parquet("./y_train.parquet")
X_val = pd.read_parquet("./X_val.parquet")
y_val = pd.read_parquet("./y_val.parquet")
X_retro = pd.read_parquet("./X_retro.parquet")
y_retro = pd.read_parquet("./y_retro.parquet")
```

```python
# Downsample to 20 samples each
AllDatasets_sample = AllDatasets.sample(n=20, random_state=42)

indices_pros = X_pros.sample(n=20, random_state=42).index
X_pros_sample = X_pros.loc[indices_pros]
y_pros_sample = y_pros.loc[indices_pros]

indices_test = X_test.sample(n=20, random_state=42).index
X_test_sample = X_test.loc[indices_test]
y_test_sample = y_test.loc[indices_test]

indices_train = X_train.sample(n=20, random_state=42).index
X_train_sample = X_train.loc[indices_train]
y_train_sample = y_train.loc[indices_train]

indices_val = X_val.sample(n=20, random_state=42).index
X_val_sample = X_val.loc[indices_val]
y_val_sample = y_val.loc[indices_val]

indices_retro = X_retro.sample(n=20, random_state=42).index
X_retro_sample = X_retro.loc[indices_retro]
y_retro_sample = y_retro.loc[indices_retro]
```

```python
AllDatasets_sample.to_parquet('AllDatasets_sample.parquet')
X_pros_sample.to_parquet('X_pros_sample.parquet')
y_pros_sample.to_parquet('y_pros_sample.parquet')
X_test_sample.to_parquet('X_test_sample.parquet')
y_test_sample.to_parquet('y_test_sample.parquet')
X_train_sample.to_parquet('X_train_sample.parquet')
y_train_sample.to_parquet('y_train_sample.parquet')
X_val_sample.to_parquet('X_val_sample.parquet')
y_val_sample.to_parquet('y_val_sample.parquet')
X_retro_sample.to_parquet('X_retro_sample.parquet')
y_retro_sample.to_parquet('y_retro_sample.parquet')
```
